# we need to outlier for each column i'e for height and weight using the following mehod.

# for height
# IQR=Q3-Q1

```python
Q1=df['Height(Inches)'].quantile(0.25)
Q3=df['Height(Inches)'].quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

# now we will going to find lower_limit and upper_limit by using the following methods

```python
lower_limit=Q1-1.5*IQR  # lower_limit formula
upper_limit=Q3+1.5*IQR  # upper_limit formula
lower_limit,upper_limit
```

# The value for (Height(Inches)) which are less than lower_limit((62.85155749999999) and greater than upper_limit(73.1257975) will be cosidered as outliers.

# using the following methods we can get outliers

```python
D1_outlier = df[(df['Height(Inches)'] < lower_limit) | (df['Height(Inches)'] > upper_limit)]
D1_outlier
```

# the above values are outliers which will remove among dataset for height

```python
D1_no_outlier= df[(df['Height(Inches)'] >lower_limit) & (df['Height(Inches)'] < upper_limit)]
D1_no_outlier
```

# second step removing outliers for weight parameters among dataset

```python
Q1=df['Weight(Pounds)'].quantile(0.25)
Q3=df['Weight(Pounds)'].quantile(0.75)
Q1,Q3
```

```python
# evaluating inter quartile range IQR for weight parameters.
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR  # lower_limit formula
upper_limit=Q3+1.5*IQR  # upper_limit formula
lower_limit,upper_limit
```

# removing outliers which has value less than lower_limit(95.93241249999997) and upper_limit(158.26911250000003))

```python
D2_outlier = df[(df['Weight(Pounds)'] < lower_limit) | (df['Weight(Pounds)'] > upper_limit)]
D2_outlier
```

# the above values are outliers which have extreme value and will be removed from the original data.

```python
# getting dataset for weight which has no outliers
D2_no_outlier= df[(df['Weight(Pounds)'] >lower_limit) & (df['Weight(Pounds)'] < upper_limit)]
D2_no_outlier
```

# check another method read the vedio

```python
import pandas as pd
import numpy as np

# Assuming df is your dataframe
# Calculate Z-scores for height and weight columns
df['height_zscore'] = (df['height'] - df['height'].mean()) / df['height'].std()
df['weight_zscore'] = (df['weight'] - df['weight'].mean()) / df['weight'].std()

# Set threshold for outliers
threshold = 3

# Identify outliers
outliers = df[(np.abs(df['height_zscore']) > threshold) | (np.abs(df['weight_zscore']) > threshold)]

# Remove outliers
df_no_outliers = df.drop(outliers.index)
```
